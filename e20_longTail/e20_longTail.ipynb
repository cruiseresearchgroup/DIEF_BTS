```python
# Sets
stALL = set(dfa['Brick'].values)
stALL = stALL.intersection(set(dfb['Brick'].values))
stALL = stALL.intersection(set(dfc['Brick'].values))
print('stALL', len(stALL))

stAC = set(dfa['Brick'].values)
stAC = stAC.intersection(set(dfc['Brick'].values))
stAC = stAC - set(dfb['Brick'].values)
print('stAC', len(stAC))

stBC = set(dfb['Brick'].values)
stBC = stBC.intersection(set(dfc['Brick'].values))
stBC = stBC - set(dfa['Brick'].values)
print('stBC', len(stBC))

stAB = set(dfa['Brick'].values)
stAB = stAB.intersection(set(dfb['Brick'].values))
stAB = stAB - set(dfc['Brick'].values)
print('stAB', len(stAB))

stA = set(dfa['Brick'].values)
stA = stA - set(dfb['Brick'].values)
stA = stA - set(dfc['Brick'].values)
print('stA', len(stA))

stB = set(dfb['Brick'].values)
stB = stB - set(dfa['Brick'].values)
stB = stB - set(dfc['Brick'].values)
print('stB', len(stB))

stC = set(dfc['Brick'].values)
stC = stC - set(dfb['Brick'].values)
stC = stC - set(dfa['Brick'].values)
print('stC', len(stC))
```

```python
vcALL = vc.loc[list(stALL)]
vcAC = vc.loc[list(stAC)]
vcBC = vc.loc[list(stBC)]
vcAB = vc.loc[list(stAB)]
vcA = vc.loc[list(stA)]
vcB = vc.loc[list(stB)]
vcC = vc.loc[list(stC)]
lvc = [vcALL,vcAB,vcAC,vcBC,vcA,vcB,vcC]
for i, ivc in enumerate(lvc):
    lvc[i] = ivc.sort_values(ascending=False)
```

```python
lsc = ['Alarm', 'Command', 'Parameter', 'Sensor', 'Setpoint', 'Status']
def get_superclass(ivc):
    r = []
    for i in ivc.index:
        sc = ''
        for j in lsc:
            if j in i:
                sc = j
        if 'Limit' in i:
            sc = 'Setpoint'
        if sc == '':
            sc = 'Other'
        r.append(sc)
    return r
        
```

```python
lsc = ['Alarm', 'Command', 'Parameter', 'Sensor', 'Setpoint', 'Status']
def get_colors(ivc):
    r = []
    for i in ivc.index:
        if 'Alarm' in i:
            r.append('C3')
        elif 'Command' in i:
            r.append('C4')
        elif 'Parameter' in i:
            r.append('C2')
        elif 'Sensor' in i:
            r.append('C0')
        elif 'Setpoint' in i:
            r.append('C8')
        elif 'Limit' in i:
            r.append('C8')
        elif 'Status' in i:
            r.append('C9')
        else:
            r.append('C7')
    return r
        
```
